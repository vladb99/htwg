```python
import random
import matplotlib.pyplot as plt
import numpy as np

f = open("data-job-length.txt", "r")
```

```python
jobFinish1, jobFinish2, jobsLength = [], [], []
for l in f:
    row = l.split()
    jobFinish1.append(row[0])
    jobFinish2.append(row[1])
    jobsLength.append(row[2])
```

```python
jobFinish1 = list(map(int, jobFinish1)) 
jobFinish2 = list(map(int, jobFinish2)) 
jobsLength = list(map(int, jobsLength))

jobFinish1 = np.array(jobFinish1)
jobFinish2 = np.array(jobFinish2)
jobsLength = np.array(jobsLength)

fairness = jobFinish1 / jobFinish2
```

```python
fig = plt.figure()
plt.plot(jobsLength, fairness, color='orange')
plt.ylim(0, 1)
plt.margins(0)

plt.xlabel('Job Length')
plt.ylabel('Unfairness')
plt.title('Figure 9.2: Lottery Fairness Study')
plt.savefig('9.2.png', dpi=227)
plt.show()
```
